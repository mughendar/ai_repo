 *** business case ***
  The term "erythemato-squamous" refers to a group of skin disorders characterized by redness (erythema) and scaling (squamous). These disorders include conditions like psoriasis, seborrheic dermatitis, lichen planus, pityriasis rosea, and more. They are often difficult to diagnose because they share similar clinical features

 ##AIM:##
   In the dataset constructed for this domain, the family history feature has the value 1 if any of these diseases has been observed in the family, and 0 otherwise. The age feature simply represents the age of the patient.

Every other feature clinical and histopathological was given a degree in the range of 0 to 3. Here, 0 indicates that the feature was not present, 3 indicates the largest amount possible, and 1, 2 indicate the relative intermediate values.

When to Use? If you believe that extreme values might skew your analysis or if your analysis requires more stable and less variable data, this approach could be beneficial.

 ***Do Not Remove Outliers (Keep All 366 Rows)***
Why? If all features are crucial for understanding skin diseases, outliers could represent rare or severe cases that are medically significant. Removing them might mean losing important insights into these conditions.
When to Use? If the analysis is exploratory, diagnostic, or aimed at understanding all possible variations within the data, keeping all rows, including outliers, might be the better choice. Especially in medical datasets, outliers can sometimes represent rare but critical cases that need to be analyzed.

```python
x=df.drop(columns=['class'])
y=df['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
Counter(y_train)
```

***since the data is highly imbalance i am going to use smote technique to balance the data***

```python
smote=SMOTE()
```

```python
x_train_s,y_train_s=smote.fit_resample(x_train,y_train)
```

```python
Counter(y_train_s)
```

 ***data has been balanced***

***scaling***

```python
sde=StandardScaler()
```

```python
x_train_scaled=sde.fit_transform(x_train_s)
x_test_scaled=sde.fit_transform(x_test)
```

***pca****

```python
pca=PCA(n_components=4)
```

```python
x_trai=pca.fit_transform(x_train_scaled)
x_tes=pca.fit_transform(x_test_scaled)
```

```python
x_trai.shape
```

```python
x_tes.shape
```

 ***implementing model svc***

```python
model=SVC()
model.fit(x_trai,y_train_s)
```

```python
predict1=model.predict(x_tes)
```

```python
accsvc=accuracy_score(y_test,predict1)
```
